```python
import os
import google.generativeai as genai


os.environ["GOOGLE_API_KEY"] = "AIzaSyDopMAB3R_Ve85_PzZ4J6vBRQUBtAaazbo"  # stays only for this kernel
# Configure API key
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))

# Create the model
model = genai.GenerativeModel("gemini-2.5-flash")


messages = []

# System instruction
system_message = input("What kind of chatbot would you like to have?\n")
messages.append({
    "role": "user","parts": [{"text": f"System instruction: {system_message}"}]
})

print("Your new assistant is ready! Type 'quit()' to exit.\n")

while True:
    user_input = input("> ")

    if user_input.strip() == "quit()":
        print("Chat ended.")
        break

    # Append user message
    messages.append({
        "role": "user",
        "parts": [{"text": user_input}]
    })

    # Generate response
    response = model.generate_content(messages)

    reply = response.text
    print("\nAssistant:", reply, "\n")

    # Save assistant reply
    messages.append({
        "role": "model",
        "parts": [{"text": reply}]
    })
```
