```python
import os
import gradio as gr
import google.generativeai as genai

# ----------------------------------------
# 1. Configure Gemini API key
# ----------------------------------------

os.environ["GOOGLE_API_KEY"] = "AIzaSyDopMAB3R_Ve85_PzZ4J6vBRQUBtAaazbo"  # stays only for this kernel


genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))

# ----------------------------------------
# 2. Conversation memory (Gemini format)
# ----------------------------------------
messages = [{
    "role": "user",
    "parts": [{"text": "System instruction: You are a financial expert specializing in real estate investment and negotiation."}]
}]

# ----------------------------------------
# 3. Chat Function
# ----------------------------------------
def CustomGemini(user_input):
    # Add user message
    messages.append({
        "role": "user",
        "parts": [{"text": user_input}]
    })

    # Gemini model call
    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(messages)

    reply = response.text

    # Add assistant reply
    messages.append({
        "role": "model",
        "parts": [{"text": reply}]
    })

    return reply

# ----------------------------------------
# 4. Gradio Interface
# ----------------------------------------
demo = gr.Interface(
    fn=CustomGemini,
    inputs="text",
    outputs="text",
    title="My Saving Assistant (Gemini)"
)

demo.launch(share=True)
```
